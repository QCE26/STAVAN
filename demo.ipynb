### Equivalance checking using Approach 1
The 2 quantum circuits are equivalent if the final tableaus are equivalent for both the circuits for |00..> and |++..> states.

```python
qc1 = STAVAN(n_qubits = 3, approach = 1)
#qc1.print_tableu_global(0)
#qc1.print_tableu_global(1)

qc1.apply_single_qubit_gate(H, 0)
qc1.apply_single_qubit_gate(H, 1)
qc1.apply_T_gate(1)
qc1.apply_CNOT_gate(0,1)
qc1.apply_CNOT_gate(1,0)
qc1.apply_CNOT_gate(0,1)
qc1.apply_CNOT_gate(0,1)
qc1.print_tableu_global(0)
qc1.print_tableu_global(1)
```

Circuits qc and qc1 are equivalent and have equivalent tableau representations.

## Approach 2 and 3
To initialize the tableau, you will have to pre-define the $T$-count and $\chi$.

```python
qc = STAVAN(n_qubits = 2, approach = 2, t_qubits = 2, chi = 4)
```

You can check and print the initial stabilizer decomposition.

```python
qc.print_tableu_global(0)
```

Apply your circuit.

```python
qc.apply_single_qubit_gate(H, 0)
qc.apply_single_qubit_gate(H, 1)
qc.apply_T_gate(0)
qc.apply_T_gate(1)
```

Estimate the output probabilites for any given output bitstream. You can use either Approach 2 or 3.

```python
#Approach 2
p00 = qc.compute_probability_Approach_2([0,1],[0,0])
p01 = qc.compute_probability_Approach_2([0,1],[0,1])
p10 = qc.compute_probability_Approach_2([0,1],[1,0])
p11 = qc.compute_probability_Approach_2([0,1],[1,1])

print(p00,p01,p10,p11)
```

```python
#Approach 3
#You need to specify the number of samples
p00 = qc.compute_probability_Approach_3([0,1], [0,0], num_samples=500)
p01 = qc.compute_probability_Approach_3([0,1], [0,1], num_samples=500)
p10 = qc.compute_probability_Approach_3([0,1], [1,0], num_samples=500)
p11 = qc.compute_probability_Approach_3([0,1], [1,1], num_samples=500)

print(p00,p01,p10,p11)
```
